```python
y = df['worldwide_gross']
x = df.drop('worldwide_gross', axis=1)
```

```python
sns.heatmap(x.corr())
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
lasso = Lasso().fit(X_train, y_train)
lasso.score(X_test, y_test)
```

```python
ridge = Ridge().fit(X_train, y_train)
ridge.score(X_test, y_test)
```

```python
from sklearn.ensemble import BaggingRegressor
```

```python
baggin = BaggingRegressor(Lasso(), n_estimators=50).fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
np.sqrt(mean_squared_error(y_test, baggin.predict(X_test)))
```

```python
np.sqrt(mean_squared_error(y_test, lasso.predict(X_test)))
```

```python
np.sqrt(mean_squared_error(y_test, ridge.predict(X_test)))
```
